## 머신러닝

### 전처리한 파일 들고와서 머신러닝

```python
encoding_df = pd.read_csv('dataset/test/ml_water_ms.csv', encoding = 'cp949', index_col = 0)
```

```python
encoding_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_features(dataDF):
    features = ['품목명', '시설유형', '재배형태']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature])
        
    return dataDF

def drop_features(df):
    df.drop(['시설유형', '재배형태'],axis=1,inplace=True)
    return df

def transform_features(df):
    df = encode_features(df)
    df = drop_features(df)
    return df
```

```python
encoding_data = encoding_df.copy()
```

```python
encoding_data = transform_features(encoding_data)
encoding_data.head()
```

### 데이터 셋 분리

```python
from sklearn.model_selection import train_test_split

X = encoding_data.drop('품목명', axis = 1)
Y = encoding_data['품목명']
seed = 11
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = seed)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 결정트리, Random Forest, 로지스틱 회귀를 위한 사이킷런 Classifier 클래스 생성
dt_clf = DecisionTreeClassifier(random_state=seed)
rf_clf = RandomForestClassifier(random_state=seed)
lr_clf = LogisticRegression(random_state=seed)

# DecisionTreeClassifier 학습/예측/평가
dt_clf.fit(X_train , y_train)
dt_pred = dt_clf.predict(X_test)
print('DecisionTreeClassifier 정확도: {0:.4f}'.format(accuracy_score(y_test, dt_pred)))

# RandomForestClassifier 학습/예측/평가
rf_clf.fit(X_train , y_train)
rf_pred = rf_clf.predict(X_test)
print('RandomForestClassifier 정확도:{0:.4f}'.format(accuracy_score(y_test, rf_pred)))

# LogisticRegression 학습/예측/평가
lr_clf.fit(X_train , y_train)
lr_pred = lr_clf.predict(X_test)
print('LogisticRegression 정확도: {0:.4f}'.format(accuracy_score(y_test, lr_pred)))
```

#### 온도 습도 반올림 자리수 1로 바꿔줌 (오차 줄이기 위함)

```python
from sklearn.preprocessing import LabelEncoder

def encode_features(dataDF):
    features = ['품목명', '시설유형', '재배형태']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature])
        
    return dataDF

def drop_features(df):
    df.drop(['시설유형', '재배형태'],axis=1,inplace=True)
    return df

def change_features(df):
    df['내부온도'] = round(df['내부온도'],1)
    df['내부습도'] = round(df['내부습도'],1)
    return df

def transform_features(df):
    df = encode_features(df)
    df = drop_features(df)
    df = change_features(df)
    return df
```

```python
encoding_data = encoding_df.copy()
```

```python
encoding_data = transform_features(encoding_data)
encoding_data.head()
```

```python
from sklearn.model_selection import train_test_split

X = encoding_data.drop('내부온도', axis = 1)
Y = encoding_data['내부온도']
seed = 11
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state = seed)
```

```python
from sklearn.metrics import *
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression


# LinearRegression 모델
lr = LinearRegression()

# 모델 훈련/피팅
lr.fit(X_train, y_train)

# 예측
pred = lr.predict(X_test)

# 모델 성능 평가
# print("정확도: ", accuracy_score(y_test, pred))
print('Mean Absolute Error: ', mean_absolute_error(y_test, pred))
print("Mean Squared Error: ", mean_squared_error(y_test, pred))
print("Mean Root Squared Error: ", np.sqrt(mean_squared_error(y_test, pred)))
```

```python
def lr_concat_pred_show(df):
    concat_df = pd.DataFrame([0])
    concat_df['machine'] = "linear Regression"
    concat_df["MAE"] = round(mean_absolute_error(y_test, pred),5)
    concat_df["MSE"] = round(mean_squared_error(y_test, pred), 5)
    concat_df["RMSE"] = round(np.sqrt(mean_squared_error(y_test, pred)), 5)
    for i in df['품목명'].unique():
        data = df[df['품목명'] == i].copy()
        concat_df[i + ": 온도 차이(<=5)비율(%)"] = round(data[data['차이(절대값)'] <= 5].shape[0] / data.shape[0] * 100, 2)
    
    concat_df = concat_df.drop([0], axis = 1)
    return concat_df
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_features(dataDF):
    features = ['품목명', '시설유형', '재배형태']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature])
        
    return dataDF

def drop_features(df):
    df.drop(['시설유형', '재배형태'],axis=1,inplace=True)
    return df

def change_features(df):
    df['내부온도'] = round(df['내부온도'],1)
    df['내부습도'] = round(df['내부습도'],1)
    return df

def transform_features(df):
    df = encode_features(df)
    df = drop_features(df)
    df = change_features(df)
    return df
```

```python
encoding_data = encoding_df.copy()
```

```python
encoding_data = transform_features(encoding_data)
encoding_data.head()
```

```python
from sklearn.model_selection import train_test_split

X = encoding_data.drop('내부온도', axis = 1)
Y = encoding_data['내부온도']
seed = 11
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state = seed)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

alphas = [0, 0.1, 1, 10, 100]
for alpha in alphas:
    ridge = Ridge(alpha = alpha)
    neg_mse_scores = cross_val_score(ridge, X, Y, scoring="neg_mean_squared_error", cv=5)
    rmse_scores = np.sqrt(-1 * neg_mse_scores)
    avg_rmse = np.mean(rmse_scores)
    
    print(f"alpha: {alpha}")
    print("5 folds의 개별 Negative MSE scores: ", np.round(neg_mse_scores, 3))
    print("5 folds의 개별 RMSE scores: ", np.round(rmse_scores, 3))
    print("5 folds의 평균 RMSE : ", np.round(avg_rmse, 3))
    print()
```
